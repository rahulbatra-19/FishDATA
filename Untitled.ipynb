```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import linear_model , preprocessing
from sklearn.linear_model import LinearRegression
```

```python
fish = pd.read_csv('Fish.csv')
fish.head(n = 11)
```

```python
plt.figure(figsize = (10,5))
plt.subplot(1,2,1)
fish.Species.value_counts().plot(kind ='bar')

```

```python
def graphs(x,fig):
    plt.subplot(4,2,fig)
    plt.title(x+ 'Histogram')
    sns.countplot(fish[x], palette = ("cubehelix"))
    
    plt.subplot(4,2,(fig+1))
    plt.title(x+ " vs Species name ")
    sns.boxplot(x= fish.Species, y = fish[x] , palette= ("cubehelix"))
```

```python
plt.figure(figsize = (20,40))
graphs('Weight',1)
graphs('Height',3)
graphs('Width',5)
plt.tight_layout()
```

```python

i = preprocessing.LabelEncoder()
fish['Species'] = i.fit_transform(fish['Species'])
fish.tail(n=14)
```

```python
fish['Length'] = (fish['Length1'] + fish['Length2'] + fish['Length3'])/3
fish['size'] = fish['Width']*fish['Height']
```

```python
print(fish)
```

```python
plt.figure(figsize=(10,15))
graphs('Length',1)
graphs('size',3)
```

```python
X = fish.drop(['Weight','Length1'],axis = 1)
# features = ['size','Length','Species']
# X= fish[features]
y = fish.Weight
```

```python
print(X)
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size =0.4,random_state= 21)
```

```python
Lr = LinearRegression()
Lr = Lr.fit(X_train, y_train)
```

```python
y_predicted= Lr.predict(X_test)
MSE = mean_squared_error(y_test, y_predicted)
r2 = r2_score(y_test, y_predicted)
```

```python
print("Mean square error is: ",MSE )
print("R2 score is: ", r2 )
```
